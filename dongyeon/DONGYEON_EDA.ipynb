```python
# join(merge)

df = pd.merge(netflix_df, ratings_df, left_on="title", right_on="Title")
df.info()
```

```python
# 별점을 찾을 수 없는 것들 제거!!!, 결측치를 찾을 수 없는 정보들 제거!!!!!
df.drop(df[df["Title_ID"] == "NOT FOUND"].index, inplace=True)
# df[(df["director"] == "NOT FOUND") & (df["cast"] == "NOT FOUND") & (df["country"] == "NOT FOUND")]
df.head()
```

```python
df.drop(["show_id", "date_added", "description","Title_ID", "Title"], axis = 1, inplace = True) # 필요없는 컬럼은 제거
```

```python
countrys = df.copy()

# 가독성을 위해 필요한 컬럼만
countrys.drop(["director", "type", "cast", "duration", "listed_in", "rating", "release_year"], axis = 1, inplace =True)
countrys.head()
```
